```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model = "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    device_map = "auto"
)

with open ("sample.txt", "r", encoding = "utf-8") as f:
    text = f.read()

prompt = f""" 
Summarize the following text in 5 bullet points:

{text}
""" 

output = pipe(
    prompt,
    max_new_tokens = 200,
    do_sample = True,
    temperature = 0.7
)

print(output[0]["generated_text"])
```
